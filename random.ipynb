```python
import sys
import requests
from bs4 import BeautifulSoup

if len(sys.argv) != 2:
    print("Usage: python script.py <URL>")
    sys.exit()

url = sys.argv[1]

headers = {
    "User-Agent": "Mozilla/5.0",
    "Accept-Language": "en-US,en;q=0.9"
}

response = requests.get(url, headers=headers)

if response.status_code != 200:
    print("Failed to fetch page:", response.status_code)
    sys.exit()

soup = BeautifulSoup(response.text, "html.parser")

# Title
title = soup.title.string.strip() if soup.title else "No Title"
print(title)

# Body
body_text = soup.get_text(separator=' ', strip=True)
print(body_text)

# Links
for link in soup.find_all("a"):
    href = link.get("href")
    if href:
        print(href)
```

```python
import sys
import urllib.request
from bs4 import BeautifulSoup

if len(sys.argv) < 2:
    print("Kindly give the url in command line")
    sys.exit()

url = sys.argv[1]

try:
    req = urllib.request.Request(url, headers={'User-Agent': 'Mozilla/5.0'})
    page = urllib.request.urlopen(req)
    html_content = page.read()

    web_page_data = BeautifulSoup(html_content, 'html.parser')

    if web_page_data.title:
        print(web_page_data.title.text.strip())
    else:
        print("No Title Found")


    print("\n")


    if web_page_data.body:
        body_text = web_page_data.body.get_text(separator=' ',strip=True)
        print(body_text)
    else:
        print("No Body Found")

    print("\n")



    all_links = web_page_data.find_all('a')
    # print(all_links)
    
    for i in range(len(all_links)):
        href = all_links[i].get('href')
        if href is not None and href != "":
            print(href)
            print("\n")
        
    for link in all_links:
        href = link.get('href')
        if href:
            print(href)
            print("\n")


except Exception as e:
    print("An error occurred:", e)
```
